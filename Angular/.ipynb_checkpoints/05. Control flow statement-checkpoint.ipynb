## Control flow statement

If you want a conditionally render something in HTML you use the same syntax as in JS, but you have to add @ symbol. 

For example you have a variable which is a boolean and you want to display paragraph which depends if it's true or false.

```python
export class ControlFlow {

  isParagraphVisible: boolean = false; ## <--- boolean on which displaying a paragraph depends on

  showP() {
    this.isParagraphVisible = true;
  }

  hideP() {
    this.isParagraphVisible = false;
  }

}

## in HTML
<table>
    <tr>
        <td>
            @if(isParagraphVisible) {    ## <--- syntax is the same as in JS but @ symbol is added
                <p>This is Sample Paragraph</p>
            } @else {
                <p>isParagraphVisible has false value</p>
            }
        </td>
        <td>
            <button type="button" (click)="showP()">Show P</button>
        </td>
        <td>
            <button type="button" (click)="hideP()">Hide P</button>
        </td>
    </tr>
</table>
```

The same for multiple conditions:

```python
<td>
    @if (monthName === "jan") {
        <span>January</span>
    } @else if (monthName === "feb") {
        <span>February</span>
    } @else if (monthName === "mar") {
        <span>March</span>
    } @else if (monthName === "apr") {
        <span>April</span>
    } @else {
        <span>Wrong input!</span>
    }
</td>
```

#### @for loops
@for loops are used to display a collection. Syntax looks like below.

```python
@for (item of items; track $index) {
    <li>{{ item.name }}</li>
    } @empty {
    <li>There are no items.</li>
}
```

The track expression is used to determine the key of the array items. In above example it's an index, but it can be also a value of object property. For example:

```python
@for (item of items; track item.name) {  ## <<--- in this case tracked key value is an object property
    <li>{{ item.name }}</li>
    } @empty {
    <li>There are no items.</li>
}
```

Below is an example of @for loop. You can display a list based on the list of items

```python
export class ControlFlow {
  cityList: string[] = ["Gda≈Ñsk", "Sopot", "Gdynia"];
  chosenCity: string = "";
}

## in HTML
<ul>
    @for (item of cityList; track $index) {
        <li>{{item}}</li>
    }
</ul>
```

or use for example in dropdown menu (pay attention to how to use ngModel in this case

```python
<select name="" id="" [(ngModel)]="chosenCity">
    @for (item of cityList; track $index) {
        <option value="{{item}}">{{item}}</option>
    }
</select>
```
