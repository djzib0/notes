## Directives

Directives add extra behaviour to HTML elements (for example routerLink instead of href in anchor tag).

#### Attributes directives

We have two common attribute directives: ngClass, and ngStyle.

##### ngClass allows to add dynamic class
##### ngStyle allows to add dynamic css


Below example of how to use ngClass in the project.
First, import NgClass in your component, and then utilize it in the HTML.


```python
import { NgClass } from '@angular/common';  ## <--- import NgClass
import { Component, signal } from '@angular/core';

@Component({
  selector: 'app-att-directive',
  imports: [NgClass],                       ## <--- import NgClass
  templateUrl: './att-directive.html',
  styleUrl: './att-directive.css',
})
export class AttDirective {

  div1ClassName = signal<string>("");

  setBgClass(className: string) {
    this.div1ClassName.set(className)
  }
}

## in HTML
<div class="row">
    <div class="col-3">
        <div class="p-3" [ngClass]="div1ClassName()"> ## <--- ngClass in square brackets, gets the signal value
            Div 1 -  {{div1ClassName()}}
        </div>
    </div>
</div>

<div class="row">
    <div class="col-3">
        <button 
            class="btn btn-success" 
            (click)="setBgClass('bg-success')" ## <--- pay attention how the argument is passed to function
        >
            Green
        </button>
        <button class="btn btn-danger" (click)="setBgClass('bg-danger')">Red</button>
    </div>
</div>
```

```python
<div class="row">
    <div class="col-3">
        <div class="p-3" [ngClass]="div1ClassName()"> ## 
            Div 1 -  {{div1ClassName()}}
        </div>
    </div>
</div>

<div class="row">
    <div class="col-3">
        <button class="btn btn-success" (click)="setBgClass('bg-success')">Green</button>
        <button class="btn btn-danger" (click)="setBgClass('bg-danger')">Red</button>
    </div>
</div>
```

ngStyle is the same as ngClass but insteaf of class name you have to use class object. Example below:

```python
import { NgClass, NgStyle } from '@angular/common';
import { Component, signal } from '@angular/core';

@Component({
  selector: 'app-att-directive',
  imports: [NgClass, NgStyle],
  templateUrl: './att-directive.html',
  styleUrl: './att-directive.css',
})
export class AttDirective {
    
  isDiv2Green: boolean = false;
    
  toggleBg() {
    this.isDiv2Green = !this.isDiv2Green;
  }

}

## in HTML

div class="col-3" >
    <div class="p-3" [ngStyle]="{'background-color': isDiv2Green === true ? 'blue' : 'red'}">
        Div - 2
    </div>
</div>
```
