## Components

Components are nothing but pages. Before adding a new component, create the "components" folder in app folder.

Enter this folder in terminal (in VSC - right click on the folder and choose "Open in integrated terminal") 

Enter command:
#### ng generate component <component name><br>
    
or
    
#### ng g c <component name><br>
    
    
A new folder "component name" will be created with necessary files. 


Every component must have a:

- TypeScript class with behaviours such as handling user fetching data from a server
- An HTML template
- CSS selector that defines how the component is used in HTML

Example below:

```python
@Component({
  selector: 'profile-photo',
  template: `<img src="profile-photo.jpg" alt="Your profile photo">`,
})
export class ProfilePhoto { }
```

To use component in the app.html use the selector name as an HTML in app.html file.

Example:

```python
<h2>Welcome</h2>
<table style="width: 100%;">
    
    <app-profile-photo></app-profile-photo>

</table>
```

You will get an error that the app-profile-photo is not a known element. It means that we have to add component class to imports in app.ts file (or in any other component file in which we would like to use this component.

```python
## app.ts

..imports

@Component({
  selector: 'app-root',
  imports: [RouterOutlet, ProfilePhoto], ## <--- here is added ProfilePhoto class
  templateUrl: './app.html',
  styleUrl: './app.css'
})
export class App {
  protected readonly title = signal('angular-tutorial');
}
```

There is a default property of components. It's called standalone property. If you want to mark the component as a not standalone, just set the property to false. Example below:

```python
@Component({
  selector: 'profile-photo',
  template: `<img src="profile-photo.jpg" alt="Your profile photo">`,
  standalone: false,  ## <--- add the property with false value
})
export class ProfilePhoto { }
```

### Adding variables to components.

To add variable in the component class, just enter the variable name (without let or const). 

```python
import { Component } from '@angular/core';

@Component({
  selector: 'app-admin',
  imports: [],
  templateUrl: './admin.html',
  styleUrl: './admin.css',
})
export class Admin {
  firstName = "Piotr";
  courseName = "Angular 21 Tutorial";
}
```

To access the variable just use double curly brackets and the variable name in the html template. It's called Angular's interpolation.

```python
<p>admin works!</p>
<h1>Admin page</h1>
<span>Admin's name is: {{firstName}}</span>
```
