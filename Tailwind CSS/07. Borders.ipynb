### Borders

###### Border Width

Syntax:<br>
class='border-{value}'

values are in pixels, so:<br>
border-0 = border-width: 0px<br>
border-2 = border-width: 2px;<br>

but if you want a border with width 1px just use <b>'border'</b>

Documentation:<br>
https://tailwindcss.com/docs/border-width

###### Border Color

Syntax:<br>
class='border-{color}-{shade}'

Documentation:<br>
https://tailwindcss.com/docs/border-color


###### Border (specific side)

Syntax:<br>
class='border-{side}-{value}'<br>
class='border-{side}-{color}-{shade}'

sides:<br>
t = top<br>
b = bottom<br>
l = left<br>
r = right<br>
x = left and right<br>
y= top and bottom<br>

###### Border Radius

Syntax:
class='rounded-{value}'

Documentation:<br>
https://tailwindcss.com/docs/border-radius




###### Border Radius (specific corner)

Syntax:<br>
class='rounded-{corner}-{value}'

Values:<br>
tl = top left<br>
tr = top right<br>
bl = bottom left<br>
br = bottom right<br>
t<br>
b<br>
l<br>
m...<br>

Documentation:<br>
https://tailwindcss.com/docs/border-radius

###### Border Style

Syntax:<br>
class='border-{value}'

Values:<br>
solid<br>
dashed<br>
dotted<br>
double<br>
hidden<br>
none<br>

Documentation:<br>
https://tailwindcss.com/docs/border-style
