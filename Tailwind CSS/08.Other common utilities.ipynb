### Other common utilities

###### Box Shadow

Syntax:<br>
class='shadow-{value}'

values:<br>
sm<br>
lg<br>
<2xl>

or single <b>shadow</b> to get a default box shadow

Documentation:<br>
https://tailwindcss.com/docs/box-shadow


###### Fill

Syntax:<br>
class='fill-none'<br>
class='fill-{color}-{shade}'<br>
class='fill-transparent'

Documentation:<br>
https://tailwindcss.com/docs/fill


###### Opacity

Syntax:<br>
class='opacity-{value}'

Documentation:<br>
https://tailwindcss.com/docs/opacity


###### Cursor

Syntax:<br>
class='cursor-{value}'

values:<br>
pointer<br>
text<br>
move<br>
not-allowed<br>

Documentation:<br>
https://tailwindcss.com/docs/cursor


###### Screen Readers

Syntax:<br>
class='[not-]sr-only'

Documentation:<br>
https://tailwindcss.com/docs/screen-readers
