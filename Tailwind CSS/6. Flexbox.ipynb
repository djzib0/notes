### Flexbox

To use flex in Tailwind add class='flex' and then another properties to align, justify and more...


###### Flex Direction

Syntax:<br>
class='flex-{direction}'

flex-row = flex-direction: row;
flex-row-reverse = flex-direction: row-reverse
flex-col = flex-direction: column;

Documentation:<br>
https://tailwindcss.com/docs/flex-direction

###### Justify Content

Syntax:<br>
class='justify-{value}'

values:
normal = normal
start = flex-start
end = flex-end
center = flex-center
between = space-between
around...
evenly...
stretch

Documentation:<br>
https://tailwindcss.com/docs/justify-content

###### Align Items

Syntax:<br>
class='item-{value}'

values:
start = flex-start
end = flex-end
center = center
baseline...
stretch...

Documentation:<br>
https://tailwindcss.com/docs/align-items

###### Gap

Syntax:<br>
class='gap-{value}'<br>
class='gap-[x/y]-{value}' <br>

Documentation:<br>
https://tailwindcss.com/docs/gap

###### Flex Grow

Have two classes
grow = flex-grow: 1
grow-0 = flex-grow: 0

Documentation:<br>
https://tailwindcss.com/docs/flex-grow

###### Flex Wrap

Syntax:<br>
class='flex-{value}'

values:
wrap = wrap
wrap-reverse = wrap-reverse
nowrap = nowrap

Documentation:<br>
https://tailwindcss.com/docs/flex-wrap
